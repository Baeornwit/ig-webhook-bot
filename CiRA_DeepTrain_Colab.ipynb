```python
import os

# ชื่อโฟลเดอร์ที่คุณอยากเช็ก (ตามรูปที่คุณส่งมา)
target_folder = "data"

if os.path.exists(target_folder):
    # สั่งให้นับเฉพาะไฟล์รูปภาพ
    all_files = os.listdir(target_folder)
    image_count = 0
    for f in all_files:
        if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
            image_count += 1

    print(f"✅ ในโฟลเดอร์ '{target_folder}' มีรูปภาพทั้งหมด: {image_count} รูป")
else:
    print(f"❌ หาโฟลเดอร์ชื่อ '{target_folder}' ไม่เจอครับ! (เช็กว่าพิมพ์ชื่อถูกไหม)")
```
